```python
import cv2
import numpy as np
import matplotlib.pylab as plt
```

```python
image = cv2.imread('images/tiger.jpg')
height, width, channels = image.shape
res = cv2.resize(image, dsize=(int(width/8), int(height/8)), interpolation=cv2.INTER_CUBIC)
height, width, channels = image.shape
height_res, width_res, channels_res = res.shape
print("\n Original Image information: \n Height: {}\n Width: {}\n Channels: {}".format(height,width,channels))
print("\n Resized Image information: \n Height: {}\n Width: {}\n Channels: {}".format(height_res,width_res,channels_res))

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
plt.figure(figsize=([12,8]))
plt.subplot(121),plt.imshow(image, vmin=0, vmax=255)
plt.title('Original Image'), 
plt.subplot(122),plt.imshow(res, vmin=0, vmax=255)
plt.title('Resized'), 
plt.show
```

```python
smaller_img = image[::2, ::2]
height_res, width_res, channels_res = smaller_img.shape
print("\n DownSampled Image information: \n Height: {}\n Width: {}\n Channels: {}".format(height_res,width_res,channels_res))
bigger_img = smaller_img.repeat(2, axis=0).repeat(2, axis=1)
height_res, width_res, channels_res = bigger_img.shape
print("\n UpSampled Image information: \n Height: {}\n Width: {}\n Channels: {}".format(height_res,width_res,channels_res))

plt.figure(figsize=([12,8]))
plt.subplot(121),plt.imshow(smaller_img, vmin=0, vmax=255)
plt.title('DownSampling'), 
plt.subplot(122),plt.imshow(bigger_img, vmin=0, vmax=255)
plt.title('UpSampling'), 
plt.show
```
